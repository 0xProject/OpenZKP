```python
def embedding_degree(p, l):
    pp = 1
    for i in range(1, 10**7):
        pp *= p
        pp %= l
        if pp == 1:
            return i
        if i % 10**6 == 0:
            print(i, pp)
    for n in range(100, 2, -1):
        if pow(p, (l-1)//n, l) == 1:
            print("(l-1/)", n)
            return (l-1)//n
    return "Good!"
```
